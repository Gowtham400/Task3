# Train Test Splitting

```python
from sklearn.preprocessing import LabelEncoder
for column in ds.columns:
    if ds[column].dtype==np.number:
        continue
    else:
        ds[column]=LabelEncoder().fit_transform(ds[column])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = ds.drop('Attrition',axis=1)
y = ds['Attrition']
```

```python
x_train, x_test , y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 0)
x_train.head()
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
```

```python
rf.fit(x_train, y_train)
```

```python
 rf.score(x_train, y_train)
```

```python
pred = rf.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

### Model accuracy with Random Forest Classifier: 85.26%

# Logistic Regression

```python
#applying Logistic Regression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
```

```python
#computing confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
confmat = confusion_matrix(y_pred, y_test)
confmat
```

```python
accuracy_score(y_pred, y_test)
```

### Model accuracy with LogisticRegression: 83.21%

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

### Model accuracy with Naive Bayes: 81.17%
